4\. load the binary file named *credit_card.dat* and convert the data into the real credit-card number.
Each line correspond to a credit card number.
Each character is composed by 6 bit (even the space) and the last 4 bit are just a padding

**hint**: use the `chr()` function to convert a number to a char

```python
filepath = "C:/Users/leonm/LaboratoryOfComputationalPhysics_Y5/"

cardnumbers = []

with open(filepath+"credit_card.dat","rb") as file:
    for line in file:
        nchar = int(len(line)//6)-1
        number = ""
        for i in [i*6 for i in range(nchar)]:
            number += chr(int(line[i:i+6],2))
        if number == "":
            continue
        cardnumbers.append(number)
print(cardnumbers)
```

5\. Load the file "user_data.json", filter the data by the "CreditCardType" field equals to "American Express". Than save the data a to CSV.

```python
filepath = "C:/Users/leonm/LaboratoryOfComputationalPhysics_Y5/"

import json
data = json.load(open("user_data.json"))

import csv

american_express = []
for user in data:
    if user["CreditCardType"] == "American Express":
        american_express.append(user)

with open(filepath+"user_data.csv", mode="w") as file:
    fieldnames = ['ID', 'JobTitle', 'EmailAddress','FirstNameLastName','CreditCard','CreditCardType']
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    for i in american_express:
        writer.writerow(i)

# with open(filepath+"user_data.csv", mode="r") as file:
#     print(file.read())
```
